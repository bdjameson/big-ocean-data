```python
from erddapy import ERDDAP
```

```python
ser = ERDDAP(server = 'https://gliders.ioos.us/erddap')
```

```python
ser.constraints =  None
ser.protocol = "tabledap"
ser.dataset_id = "ce_320-20220720T1429"
opendap_url = ser.get_download_url(
    response="opendap",
)
print(opendap_url)
```

```python
ser.response = "nc"
```

```python
df1 = ser.to_pandas(
    index_col="time (UTC)",
    parse_dates=True,
)
```

```python
df1.rename(columns={df1.columns[3]: 'Lat',df1.columns[4]: 'Long',df1.columns[5]: 'Depth',df1.columns[6]: 'Scatter',
                   df1.columns[7]: 'CDOM',df1.columns[8]: 'Chl', df1.columns[9]: 'Conductivity'},inplace=True)
```

```python
start320 = '2022-08-03 20:05:54+00:00'
end320 = '2022-08-07 11:41:59+00:00'

#start383 = '2022-08-06 00:00:00+00:00'
#end383 = '2022-08-11 19:54:11+00:00'
```

```python
df = df1.loc[start320:end320,:]
df.head(2)
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

import matplotlib.colors as mcolors
import matplotlib.cm
```

```python
plt.plot(df.index, df['Depth'])
plt.gca().invert_yaxis()
```

```python
fig, ax1 = plt.subplots() 
ax1.set_xlabel('UTC TIME') 
ax1.set_ylabel('Depth (m)') 
sp=ax1.scatter(df.index, df['Depth'], c=df['temperature (Celsius)'], s=50, marker='_')

plt.gca().invert_yaxis()
cbar = fig.colorbar(sp, ax=ax1)
cbar.set_label('Temperature ($^\circ$C)', rotation=270, labelpad=20)
```

```python
fig, ax1 = plt.subplots() 
ax1.set_xlabel('UTC TIME') 
ax1.set_ylabel('Depth (m)') 
sp=ax1.scatter(df.index, df['Depth'], c=df['dissolved_oxygen (umol kg-1)'], 
            s=50, marker='_')

plt.gca().invert_yaxis()
cbar = fig.colorbar(sp, ax=ax1)
cbar.set_label('Dissolved ' r'$O_2$ ' r'($\mu mol$ ' r'$kg^{-1}$)', rotation=270, labelpad=20)
```
