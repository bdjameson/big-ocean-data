```python
from erddapy import ERDDAP
```

```python
ser = ERDDAP(server = 'https://gliders.ioos.us/erddap')
```

```python
ser.constraints =  None
ser.protocol = "tabledap"
ser.dataset_id = "ce_320-20220720T1429"
opendap_url = ser.get_download_url(
    response="opendap",
)
print(opendap_url)
```

```python
ser.response = "nc"
```

```python
df1 = ser.to_pandas(
    index_col="time (UTC)",
    parse_dates=True,
)
```

```python
df1.rename(columns={df1.columns[3]: 'Lat',df1.columns[4]: 'Long',df1.columns[5]: 'Depth',df1.columns[6]: 'Scatter',
                   df1.columns[7]: 'CDOM',df1.columns[8]: 'Chl', df1.columns[9]: 'Conductivity'},inplace=True)
```
